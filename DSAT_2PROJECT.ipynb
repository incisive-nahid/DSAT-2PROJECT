## Modules and packages used

NUMPY:
It is used to plot image so that it is visible to user.

     fliplr()-B = fliplr(A) returns A with its columns flipped in the left-right direction (that is, about a vertical axis)

     flipud()-B = flipud(A) returns A with its rows flipped in the up-down direction (that is, about a horizontal axis).

Matplotlib.pylab: 

     imread()-It is used to load an image in the Python program from the specified file.

     imshow()-The imshow() function in pyplot module of matplotlib library is used to display data as an image; i.e. on a 2D          regular raster

SKIMAGE:

    rgb2gray()-The rgb2gray function converts RGB images to grayscale by eliminating the hue and saturation information 
    while retaining the luminance. 

# FLOWCHART

```python
import numpy as np
import matplotlib.pylab as plt

from numpy import fliplr, flipud
from skimage.color import rgb2gray


def LoadImage():
    path = str(input("Enter path : "))
    img_name = str(input("Enter file name : "))
    target = path+"\\"+img_name+".jpg"  # user entered path and file name

    image = plt.imread(target)  # image contains 3d array
    return image


def displayImage():
    image = LoadImage()
    imgplot = plt.imshow(image)
    plt.show()  # to view the image


def EditImage():
    choice = int(input('''Press
    1. Horizontal Flip
    2. Vertical Flip
    3. Invert Colors
    4. Convert to Grayscale
    5. Convert to B&W\n'''))

    if (choice == 1):
        image = LoadImage()  # Original image
        # pre-defined function to horizontally flipped image
        image_hf = np.fliplr(image)

        plt.subplot(121)
        plt.imshow(image)
        plt.title('Original Image')  # printing oiginal image

        plt.subplot(122)
        plt.imshow(image_hf)
        # printing horizontally flipped image
        plt.title('Horizontally\n flipped \nImage')

        plt.show()
        return image_hf

    elif (choice == 2):
        image = LoadImage()

        # pre-defined function to vertically flip image
        image_vf = flipud(image)
        plt.subplot(121)
        plt.imshow(image)
        plt.title('Original Image')  # printing original image

        plt.subplot(122)
        plt.imshow(image_vf)  # printing vertically flipped image
        plt.title('Vertically\n Flipped\n Image')

        plt.show()
        return image_vf

    elif (choice == 3):
        image = LoadImage()
        # to invert the color of image using pre-defined invert function
        image_inv = np.invert(image)
        plt.subplot(121)
        plt.imshow(image)
        plt.title("Original Image")

        plt.subplot(122)
        plt.imshow(image_inv)
        plt.title("Inverted Image")

        plt.show()
        return image_inv

    elif (choice == 4):
        image = LoadImage()
        # to convert image to grayscale using pre-defined rgb2gray function
        new_image = rgb2gray(image)

        plt.subplot(121)
        plt.imshow(image)
        plt.title('Original Image')

        plt.subplot(122)
        plt.imshow(new_image)
        plt.title('Grayscale Format')

        plt.show()
        return new_image

    elif (choice == 5):
        image = LoadImage()
        new_img = rgb2gray(image)
        plt.subplot(121)
        plt.imshow(image)
        plt.title("Original Image")

        plt.subplot(122)
        # first converting original image to grayscale and then to B&W
        plt.imshow(new_img, cmap=plt.cm.gray)
        plt.title("B&W Image")

        plt.show()
        return new_img

    else:
        print("\n**INVALID CHOICE**\n")
        EditImage()


def SaveImage():
    # image=LoadImage()
    imgplot = plt.imshow(image)
    # plt.show()
    # new location where the edited image is to be saved
    path = str(input("Enter path new path where image is to be stored : "))
    n_img_name = str(input("Enter new file name of image : ")
                     )  # name of the new file
    target = path+"\\"+n_img_name
    # to convert normal string to raw string siffix r is used
    raw_s = r'{}'.format(target)
    plt.savefig(raw_s)


while (True):
    ch = (input('''Enter your choice
    1.Load image
    2.Edit image
    3.Save image
    4.Quit\n'''))
    if (ch == '1'):
        displayImage()
    elif (ch == '2'):
        image = EditImage()

    elif (ch == '3'):
        print("*Saving Image*")
        SaveImage()
    elif (ch == '4'):
        print("\n\t -----Exiting the editor-----\t\n")
        break
    else:
        print("\n**INVALID CHOICE**\n")
```
