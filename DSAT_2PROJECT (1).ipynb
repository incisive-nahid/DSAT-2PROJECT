```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io



def LoadImage():
    path = str(input("Enter path : "))
    img_name = str(input("Enter file name : "))
    target = path+"\\"+img_name+".jpg"  # user entered path and file name

    image = io.imread(target)  # image contains 3d array
    return image


def displayImage():
    image = LoadImage()
    imgplot = plt.imshow(image)
    plt.show()  # to view the image


def EditImage():
    choice = int(input('''Press
    1. Horizontal Flip
    2. Vertical Flip
    3. Invert Colors
    4. Convert to Grayscale
    5. Convert to B&W\n'''))

    if (choice == 1):
        image = LoadImage()  # Original image
        # pre-defined function to horizontally flipped image
        image_hf = image[:,::-1]
        plt.imshow(image)
        plt.title('Original Image')
        
        plt.imshow(image_hf)
        plt.title('Horizontally\n flipped \nImage')
        plt.show()
        return image_hf
        
        

    elif (choice == 2):
        image = LoadImage()

        # pre-defined function to vertically flip image
        image_vf = image[::-1]
        plt.imshow(image)
        plt.title('Original Image')  # printing original image


        plt.imshow(image_vf)  # printing vertically flipped image
        plt.title('Vertically\n Flipped\n Image')

        plt.show()
        return image_vf

    elif (choice == 3):
        image = LoadImage()
        # to invert the color of image using pre-defined invert function
        image_inv = image[:,:,::-1]
 
        plt.imshow(image)
        plt.title("Original Image")


        plt.imshow(image_inv)
        plt.title("Inverted Image")

        plt.show()
        return image_inv

    elif (choice == 4):
        image = LoadImage()
        # to convert image to grayscale using pre-defined rgb2gray function
        new_image = np.mean(image,axis=-1)

        plt.imshow(image)
        plt.title('Original Image')

        plt.imshow(new_image,cmap="Greys")
        plt.title('Grayscale Format')

        plt.show()
        return new_image

    elif (choice == 5):
        image = LoadImage()
        new_image = np.mean(image,axis=-1)

        plt.imshow(image)
        plt.title("Original Image")

        b_w = np.copy(new_image)
        b_w[b_w < 128] = 0
        b_w[b_w >= 128] = 255
        plt.imshow(b_W, cmap="Greys")           # converting image to black and white
        plt.title("Black &\n White\n Image: ")
        plt.show()
        return b_w

    else:
        print("\n**INVALID CHOICE**\n")
        EditImage()


def SaveImage():
    # image=LoadImage()
    imgplot = plt.imshow(image)
    # plt.show()
    # new location where the edited image is to be saved
    path = str(input("Enter path new path where image is to be stored : "))
    n_img_name = str(input("Enter new file name of image : ")
                     )  # name of the new file
    target = path+"\\"+n_img_name
    # to convert normal string to raw string siffix r is used
    raw_s = r'{}'.format(target)
    plt.savefig(raw_s)


while (True):
    ch = (input('''Enter your choice
    1.Load image
    2.Edit image
    3.Save image
    4.Quit\n'''))
    if (ch == '1'):
        displayImage()
    elif (ch == '2'):
        image = EditImage()

    elif (ch == '3'):
        print("*Saving Image*")
        SaveImage()
    elif (ch == '4'):
        print("\n\t -----Exiting the editor-----\t\n")
        break
    else:
        print("\n**INVALID CHOICE**\n")
```
